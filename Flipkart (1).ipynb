### Remove nan values in Price

```python
plt.figure(figsize = [5,3])
sns.distplot(df['Price'])
```

```python
median_Price = df['Price'].median()
df['Price'].fillna(median_Price,inplace = True)
df.isnull().sum()
```

### Remove nan values in Discount_Percentage

```python
plt.figure(figsize = [5,3])
sns.distplot(df['Discount_Percentage'])
```

```python
median_Discount_Percentage = df['Discount_Percentage'].median()
df['Discount_Percentage'].fillna(median_Discount_Percentage,inplace = True)
df.isnull().sum()
```

```python
df.dropna(subset=["Stock_Availability"], inplace=True)
```

```python
df.isnull().sum()
```

### Remove nan values in Rating

```python
plt.figure(figsize = [5,3])
sns.distplot(df['Rating'])
```

```python
median_Rating = df['Rating'].median()
df['Rating'].fillna(median_Rating,inplace = True)
df.isnull().sum()
```

## Data Visualization

### Q1. Which category has the highest number of products, and how could this influence Flipkart’s marketing or inventory strategies?

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, y="Category", order=df["Category"].value_counts().index,hue = "Category")
plt.title("Product Count by Category")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```

### Q2. What is the most common price range for products, and how can Flipkart use this information to adjust pricing strategies or promotional efforts?

```python
plt.figure(figsize=(6, 4))
sns.histplot(df["Price"], kde=True, bins=20)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

### Q3. What percentage of products are out of stock, and how can Flipkart address stock availability issues to improve customer satisfaction and sales?

```python
stock_counts = df["Stock_Availability"].value_counts()
plt.figure(figsize=(6, 4))
plt.pie(stock_counts, labels=stock_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Stock Availability Distribution")
plt.show()
```

### Q4. How does price impact product ratings across different categories, and what pricing strategies can Flipkart implement to enhance customer satisfaction?

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x="Price", y="Rating", hue="Category", alpha=0.7)
plt.title("Price vs Rating by Category")
plt.xlabel("Price")
plt.ylabel("Rating")
plt.show()
```

### Q5. Which category shows the highest rating variation, and how can Flipkart use this insight to improve product quality and customer experience?

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x="Category", y="Rating")
plt.xticks(rotation=45)
plt.title("Rating Distribution by Category")
plt.show()
```

### Q6. What is the correlation between product features (Price, Discount Percentage, Rating, and Number of Reviews), and how can Flipkart use these insights to improve its business strategy?

```python
numerical_columns = df[["Price", "Discount_Percentage", "Rating", "Number_of_Reviews"]].dropna()
plt.figure(figsize=(6, 4))
sns.heatmap(numerical_columns.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### Q7. How does the price distribution vary across different product categories on Flipkart, and what insights can be derived to optimize pricing strategies for different segments?

```python
plt.figure(figsize=(6, 4))
sns.violinplot(data=df, x="Category", y="Price", inner="quartile")
plt.xticks(rotation=45)
plt.title("Price Distribution by Category")
plt.show()
```

### Q8. How does stock availability impact the total number of reviews across different product categories on Flipkart, and what actions can be taken to manage inventory effectively for highly reviewed categories?

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, y="Category", x="Number_of_Reviews", estimator=sum, ci=None,hue = "Stock_Availability")
plt.title("Total Number of Reviews vs Category by Stock_Availability")
plt.xlabel("Total Reviews")
plt.ylabel("Category")
plt.show()
```

### Q9.How can the distribution of shipping charges on Flipkart be optimized to improve customer satisfaction, considering the concentration of charges around certain price points?

```python
plt.figure(figsize=(6, 4))
sns.histplot(df["Shipping_Charges"], kde=True, bins=20, color="green")
plt.title("Distribution of Shipping Charges")
plt.xlabel("Shipping Charges")
plt.ylabel("Frequency")
plt.show()
```

### Q10. How can Flipkart leverage the category-wise average discount information to maximize customer engagement and conversion rates?

```python
avg_discount_category = df.groupby("Category")["Discount_Percentage"].mean().sort_values(ascending=False)
avg_discount_category.plot(kind="barh", figsize=(6, 4), color="Red")
plt.title("Average Discount Percentage by Category")
plt.xlabel("Average Discount Percentage")
plt.ylabel("Category")
plt.show()
```
